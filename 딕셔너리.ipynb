`update` 메서드를 사용하면 딕셔너리 두 개를 합칠 수도 있습니다.

`key`가 문자라면 `key1=value1, key2=value2` 형태로 `update` 메서드를 사용할 수 있습니다. 

### 딕셔너리 삭제

딕셔너리의 데이터는 del 키워드를 사용해서 삭제할 수 있습니다.

`pop`메서드는 `key`를 입력받아 그 값을 반환합니다.   
데이터는 딕셔너리에서 삭제됩니다. 

`popitem` 메서드는 가장 마지막에 저장된 `key`와 `value` 한 쌍을 튜플로 반환합니다. 

`clear`는 전체 데이터를 삭제합니다.
